```python
import pandas as pd
```

```python
data = pd.read_csv('cleaned_data')
data.head()
```

```python
a = data['ted_types']
```

```python
cat_rating= a.astype('category')
onedummycolumn = cat_rating.cat.codes
dummyted = pd.get_dummies(cat_rating)
```

```python
b = data['top_rating']
```

```python
cat_rating= b.astype('category')
onedummycolumn = cat_rating.cat.codes
dummyrating = pd.get_dummies(cat_rating)
```

```python
newdata = pd.concat([data, dummyted, dummyrating], axis=1)
```

```python
pd.set_option('display.max_columns', 100)
```

```python
realdata = newdata.drop(['Unnamed: 0', 'ted_types', 'top_rating', 'description'], 1)
realrealdata = realdata.rename(columns={'Other TED': 'Other_TED', 'Jaw-dropping': 'Jaw_dropping'})
realrealdata.head()
```

```python
realrealdata['views'] = realrealdata['total_views']
momma = realrealdata.drop(columns = 'total_views')
momma
```

```python
from statsmodels.formula.api import ols
import statsmodels.api as sm
```

```python
outcome = 'total_views'
predictors = realrealdata.drop('total_views', axis=1)
pred_sum = "+".join(predictors.columns)
formula = outcome + "~" + pred_sum
```

```python
model = ols(formula= formula, data= realrealdata).fit()
model.summary()
```

```python
from sklearn import svm
from sklearn.datasets import samples_generator
from sklearn.feature_selection import SelectKBest, f_regression
from sklearn.pipeline import make_pipeline
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report

print(__doc__)
```

```python
X = momma.iloc[:, :-1]
y = momma['views']
```

```python
# import some data to play with
X, y = samples_generator.make_classification()

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)

# ANOVA SVM-C
# 1) anova filter, take 3 best ranked features
anova_filter = SelectKBest(f_regression, k=6)
# 2) svm
clf = svm.SVC(kernel='linear')

anova_svm = make_pipeline(anova_filter, clf)
anova_svm.fit(X_train, y_train)
y_pred = anova_svm.predict(X_test)
print(classification_report(y_test, y_pred))
```
